```python
# step3: Working with Text Data
#Extract First and Last Names
import pandas as pd
df = pd.read_csv("mega_employee_sales.csv")
df['full_name'] = df['FullName'].astype(str)
df['first_name'] = df['FullName'].str.split().str[0]
df['last_name'] = df['FullName'].str.split().str[1]
print(df['first_name'].head(5))
print(df['last_name'].head(5))
```

```python
# create company email
df['email'] = (df['first_name'].str.lower()+'.'+df['last_name'].str.lower()+'@company.com')
print(df['email'].head(5))
```

```python
#Step4: Simple Filters
# Employees with salary > 100000

df[df['Salary'] > 10000]
```

```python
# Sales department only
df[df['Department'] == 'Sales']
```

```python
# Multiple Condititons
# Sales employees in India with salary > 80000

df[(df['Department'] == 'Sales') & (df['Country']=='India') & (df['Salary'] > 8000)]
```

```python
#Step5: Date Handling
# Convert Columns to Date Format
import pandas as pd
df = pd.read_csv("mega_employee_sales.csv")
df.columns= df.columns.str.strip()
df['Joining_Date']=pd.to_datetime(df['Joining_Date'],errors='coerce')
```

```python
print(df.head(5))
```

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'],errors = 'coerce')
print(df.head(5))
```

```python
#Extract Date Parts
df['Joining_Year']=df['Joining_Date'].dt.year
df['Joining_Month']=df['Sale_Date'].dt.month
print(df.head(5))
```

```python
#Filter by Date Range
# Sales in 2024
df['Sale_2024']= df['Sale_Date'].dt.year==2024
print(df['Sale_2024'].head(5))
```

```python
#Count of 2024 Sales
count_2024 = df['Sale_2024'].sum()
print("Number of sales in 2024:", count_2024)
```

```python
#Step 6. Calculated Columns
#Total Revenue
df['Total_Sale'] = df['Quantity']*df['Price']
print(df['Total_Sale'].head(5))
```

```python
#Performance Bonus
#Example: 10% of salary for "Excellent" performers.
import numpy as np
df['Bonus']=np.where(df['Performance']=='Excellent',df['Salary']*0.10,0)
df
```

```python
##Step 7. GroupBy & Aggregations
#Average Salary by Department
print(df.groupby('Department')['Salary'].mean())
```

```python
#Total Sales by Country
print(df.groupby('Country')['Total_Sale'].sum())
```

```python
#Multiple Metrics
df.groupby('Department').agg({
    'Salary': ['mean', 'max'],
    'Total_Sale': 'sum'
})
```
